```r
##This notebook uses Bioconductor packages:
#- AnnotationHub
#- ensembldb
```

```r
library(ensembldb)
library(AnnotationHub)

ah <- AnnotationHub()
edb <- ah[["AH119325"]]
```

```r
gene <- genes(edb, filter = ~ gene_name == "GHR")
gene
```

```r
tx <- transcripts(edb, filter = ~ gene_name == "GHR")
tx

write.table(tx,
            file = "GHR_transcripts.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```

```r
prot <- proteins(edb, filter = ~ gene_name == "GHR")
prot

write.table(prot,
            file = "GHR_proteins.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```

```r
tx_ids <- prot$tx_id
tx_ids

tx_gr <- transcripts(edb, filter = ~ tx_id %in% tx_ids)
tx_gr

mapping_df <- as.data.frame(tx_gr)

write.table(mapping_df,
            file = "GHR_protein_genome_map.tsv",
            sep = "\t",
            row.names = FALSE,
            quote = FALSE)
```
